 ### Age distribution all participants

```python
# Please select your age.
count = df['Please select your age.'].value_counts()

results = pd.DataFrame(count.values, count.index)
results.columns = ['Age category']
display(results)
```

```python
order=['18 - 24','25 - 34','35 - 44', '45 - 54', '55 or older']
ax = plt.figure(figsize=(10,6))
count = df['Please select your age.'].value_counts()

sns.set(style="darkgrid")
sns.barplot(count.index, count.values, order=order)
plt.xticks(rotation= 90)
plt.title('Please select your age.', bbox={'facecolor':'0.8', 'pad':12})
plt.ylabel('Number of participants', fontsize=15)
plt.xlabel('Age category', fontsize=15)

ax = plt.gca()
totals = []
for i in ax.patches:
    totals.append(i.get_height())
    
total = sum(totals)

for i in ax.patches:
    ax.text(i.get_x()+ 0.15, i.get_height()+.9, \
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=17,
                color='dimgrey')

#plt.show()
display(plt)
```

### Do you write code as part of your job?

```python
# Do you write code as part of your job?
count = df['Do you write code as part of your job?'].value_counts()
results = pd.DataFrame(count.values, count.index)
results.columns = ['Do you write code as part of your job?']
display(results)
```

```python
# Pie Chart software development Yes/No
df.groupby('Do you write code as part of your job?')['Do you write code as part of your job?'].count().plot.pie( autopct='%1.1f%%',figsize=(6,6), fontsize = 20)
count = df['Do you write code as part of your job?'].value_counts()

plt.title('Do you write code as part of your job?', bbox={'facecolor':'0.8', 'pad':12})
plt.show()
```

```python
# Relationship age and write_code -> nur alter write_code yes
order=['18 - 24','25 - 34','35 - 44', '45 - 54', '55 or older']
plt.figure(figsize=(13,8))
sns.countplot(x='Please select your age.', hue='Do you write code as part of your job?', data=df, palette="Greens_d", order=order)
plt.title('Relation between age and write code ', bbox={'facecolor':'0.8', 'pad':12})
plt.ylabel('Number of participants', fontsize=15)
plt.xlabel('Age category', fontsize=15)    
plt.show()
```
