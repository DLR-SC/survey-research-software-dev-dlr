## When do you test your software?

```python
label_text=['never','sometimes','regularly']
fig, ax =plt.subplots(1,3, sharey=True, figsize=(20,8))
ax4 = sns.countplot(df['Before I publish results based on it'], ax=ax[2])
ax4.set_xticklabels(label_text)
ax4.set_title("Before I publish results based on it")
ax5 = sns.countplot(df['Before I give it to someone else'], ax=ax[1])
ax5.set_xticklabels(label_text)
ax5.set_title("Before I give it to someone else")
ax6 = sns.countplot(df['During development'], ax=ax[0])
ax6.set_xticklabels(label_text)
ax6.set_title("During development")
fig.suptitle('When do you test your software?', bbox={'facecolor':'0.8', 'pad':12})
plt.show()
```

##  How do you test your software?

```python
fig, ax =plt.subplots(1,3, sharey=True, figsize=(20,8))
ax1 = sns.countplot(df['Automated'], ax=ax[2])
ax1.set_xticklabels(label_text)
ax1.set_title("Automated")
ax2 = sns.countplot(df['semi-automated'], ax=ax[1])
ax2.set_xticklabels(label_text)
ax2.set_title("Semi-automated")
ax3 = sns.countplot(df['manually'], ax=ax[0])
ax3.set_xticklabels(label_text)
ax3.set_title("Manually")
fig.suptitle('How do you test your software?', bbox={'facecolor':'0.8', 'pad':12})
plt.show()


```
